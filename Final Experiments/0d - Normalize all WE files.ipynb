```python
labels = [m['label'] for m in models]
```

As can be seen below the values of dimension can have a great variance (example is for `FlauBERT small`).
Notably, dimension 52 has significantly lower values for words for the sample of words, while dimension 44 has higher values.

Therefore, we perform min max normalization of dimensions prior to any other experiments.

# Adjs

```python
for label in labels:
    df = pd.read_csv(f'../Data/{label}/all_adjs_we.csv', index_col=0)
    df.iloc[:, :-3] = (df.iloc[:, :-3]-df.iloc[:, :-3].min())/(df.iloc[:, :-3].max()-df.iloc[:, :-3].min())
    df.to_csv(f'../Data/{label}/all_adjs_we.csv')
```
