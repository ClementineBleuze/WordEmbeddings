# Dimension clustering and evaluation

As seen above, there is a significant overlap for all observed dimensions for values of nouns and verbs, however, visually it is difficult to compare overlap of different dimensions.

As our next step, we can attempt to use K-means clustering for all the dimensions above and create 2 clusters. Then for each obtained clustering we will compute **Precision**, **Recall** and **F-Measure** and use these metrics to compare how well a selected dimension is for predicting noun-ness/verb-ness.

We will start with the dimensions above: `159`, `192`, `198`, `282`, `29`, `310`, `341`, `401`, `462`, `480`.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import precision_recall_fscore_support
```

```python
# The original PoS tags
y_true = unique_n_v['pos']
```

```python
dimensions = ['159', '192', '198', '282', '29', '310', '341', '401', '462', '480']
```

```python
import sys
sys.path.append('../Util')

from clustering import *
```

```python
dimension_clustering_df(y_true, dimensions, unique_n_v)
```

We can see that precision, recall and F-measure fro dimensions `480` and `462` is the highest and is very close to one another, with it being closely followed by dimension `159` and a slightly bigger gap with other dimensions. This exactly replicates the results of correlation experiment https://github.com/ClementineBleuze/WordEmbeddings/blob/main/FlauBERT/pos_correlation_experiment_v1.ipynb where dimensions `480` and `462` had the highest correlation to PoS and had very close values and were followed by `159` with a slightly bigger gap.

We can make an assumption that correlation test and Precision, Recall, F-measure of clustering convey the same information.

# Dimension combinations

Now that we looked at Precision/Recall/F-measure for one dimension, it would be interesting to compare the results if 2 dimensions are taken together. As we saw in the correlation experiment, even though dimensions `480` and `462` have high correlation to PoS information for verbs and nouns, between the dimensions themselves the correlation is relatively low.

This leading to hypothesis that a certain pair of dimensions can encode PoS information better than just one dimension.

To measure this we will repeat the experiment above, using all possible pairs of dimensions and obtain the same metrics as above and sort them similarly.

```python
from itertools import combinations
```

```python
dim_combinations = list(combinations(unique_n_v.columns[:-6], 2))
```

```python
dim_combinations_df = dimension_clustering_df(y_true, dim_combinations, unique_n_v, two_dim=True, logs=True)
```

```python
dim_combinations_df.sort_values(by='F_macro_avg', ascending=False)[:10]
```

We can see indeed that combining information from 2 dimensions seems to be more precise in classifying Nouns vs Verbs.
At the top of the list of dimensions with the highest F-measure we mainly see combinations of previously inspected dimensions. At least one of the dimensions is in each pair is from top 3 dimensions from both experiment with highest correlation and with highest F-measure/Precision/Recall for one dimension.

We can see dimensions like `128` and `375` appear on top, even though they were not previously seen in other test.

In general we can try to loosely categorize the outliers into different categories:
- Can be interpreted as a verb
- Can be interpretted as an adverb
- Can be interpretted as an interjection
- Can be interpretted as adjective

We can add this data based on Morphalou's data and based on the observations above.

```python
manual_adv = ['lundi', 'mardi', 'mercredi', 'jeudi', 'vendredi', 'samedi', 'dimanche', 
              'soir', 'matin', 'lendemain', 'demi-heure', 
              'fois', 'moment', 'instant', 'habitude', 'époque', 'présent', 'coup', 'suite']

manual_intj = ['félicitations', 'dommage', 'mesdames', 'madame', 'mademoiselle', 'problème', 'souci']

manual_verb = ['chant', 'cour', 'sort', 'sorte']
```

```python
outlier_nouns.loc[:, "Verb_like"] = False
outlier_nouns.loc[:, "Adverb_like"] = False
outlier_nouns.loc[:, "Adjective_like"] = False
outlier_nouns.loc[:, "Interjection_like"] = False
outlier_nouns.loc[:, "GW_like"] = False
```

```python
outlier_nouns.loc['tout']['total_pos_list']
```

```python
for w in outlier_nouns.index:
    pos_list = outlier_nouns.loc[w]['total_pos_list']
    if 'Verb' in pos_list or w in manual_verb:
        outlier_nouns.loc[w, 'Verb_like'] = True
    if 'Interjection' in pos_list or w in manual_intj:
        outlier_nouns.loc[w, 'Interjection_like'] = True
    if 'Adverb' in pos_list or w in manual_adv:
        outlier_nouns.loc[w, 'Adverb_like'] = True
    if 'Adjective' in pos_list:
        outlier_nouns.loc[w, "Adjective_like"] = True
    if 'GW' in pos_list:
        outlier_nouns.loc[w, "GW_like"] = True
 
```

```python
outlier_nouns['pos_count'] = outlier_nouns.loc[:, ['Verb_like', 'Adverb_like', 'Adjective_like', 'Interjection_like', 'GW_like']].sum(axis=1) + 1
outlier_nouns
```

```python
sns.histplot(outlier_nouns, x='pos_count', bins=4)
```
