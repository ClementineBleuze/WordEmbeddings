# Multiple correlation

We can test a theory that frequency affects the correlation of dimensions with PoS using [multiple correlation](https://real-statistics.com/correlation/multiple-correlation/). We will test whether a dimension is correlated with both PoS information and frequency.

```python
cor = unique_n_v.corr()
```

```python
import math
```

```python
# Independent variables
x = 'pos'
y = 'freq'

n = len(unique_n_v)

dim_results = []

# Dependent variable
for z in range(512):
    z = str(z)

    # Pairings
    xz = cor.loc[ x, z ]
    yz = cor.loc[ y, z ]
    xy = cor.loc[ x, y ]

    Rxyz = math.sqrt((abs(xz**2) + abs(yz**2) - 2*xz*yz*xy) / (1-abs(xy**2)) )
    R2 = Rxyz**2

    # Calculate adjusted R-squared
     # Number of rows
    k = 2       # Number of independent variables
    R2_adj = 1 - ( ((1-R2)*(n-1)) / (n-k-1) )
    
    dim_res = {'Dim': z, 'R2': R2, 'R2_adj': R2_adj}
    dim_results.append(dim_res)
```

```python
pd.DataFrame(dim_results).sort_values(by='R2', ascending=False)[:10]
```
